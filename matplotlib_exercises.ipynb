1. Use matplotlib to plot the following equation:

y = x^2 âˆ’ x + 2
You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, as the origin.

```python
x = np.linspace(-5, 5, 100)

y = x ** 2 - x + 2

plt.plot(x, y, label='y = x**2 - x +2')

plt.annotate('Orgin\n(0. 0)', xy=(0, 0), xytext=(1, 10),
             textcoords='offset point', arrowprops=dict(arrowstyle='->'))

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = x**2 - x +2')

plt.legend()

plt.grid(True)
plt.show()
```

Create and label 4 separate charts with the following equations (choose a range for x that makes sense):

y=âˆšx
y=x**3
y=2**x
y=1/(x+1)


```python
fig, axs = plt.subplots()

# Equation 1: y = âˆšx
x1 = np.linspace(0, 10, 100)
y1 = [math.sqrt(xi) for xi in x1]

axs.plot(x1, y1)
axs.set_title('y=âˆšx')
axs.set_xlabel('x')
axs.set_ylabel('y')
```

```python
fig, axs = plt.subplots()

x2 = np.linspace(-2, 2, 100)
y2 = [math.pow(xi, 3) for xi in x2]
axs.plot(x2, y2)
axs.set_title('y=x^3')
axs.set_xlabel('x')
axs.set_ylabel('y')
```

```python
fig, axs = plt.subplots()

x3 = np.linspace(-1, 3, 100)
y3 = [math.pow(2, xi) for xi in x2]
axs.plot(x3, y3)
axs.set_title('y=2^x')
axs.set_xlabel('x')
axs.set_ylabel('y')
```

```python
fig, axs = plt.subplots()

x4 = np.linspace(-2, 2, 100)
y4 = [1 / (xi + 1) for xi in x4]
axs.plot(x4, y4)
axs.set_title('y=1/(x+1)')
axs.set_xlabel('x')
axs.set_ylabel('y')
```

Combine the figures you created in the last step into one large figure with 4 subplots.

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Equation 1: y = âˆšx
x1 = np.linspace(0, 10, 100)
y1 = [math.sqrt(xi) for xi in x1]

axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('y=âˆšx')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')

x2 = np.linspace(-2, 2, 100)
y2 = [math.pow(xi, 3) for xi in x2]
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('y=x^3')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('y')

x3 = np.linspace(-1, 3, 100)
y3 = [math.pow(2, xi) for xi in x2]
axs[1,0].plot(x3, y3)
axs[1,0].set_title('y=2^x')
axs[1,0].set_xlabel('x')
axs[1,0].set_ylabel('y')

x4 = np.linspace(-2, 2, 100)
y4 = [1 / (xi + 1) for xi in x4]
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('y=1/(x+1)')
axs[1,1].set_xlabel('x')
axs[1,1].set_ylabel('y')
```
