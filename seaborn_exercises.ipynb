5. Load the sleepstudy dataset from pydataset and read it's documentation. Use seaborn to create a line chart of all the individual subject's reaction times and a more prominant line showing the average change in reaction time.

```python
df = data('sleepstudy')
df
```

```python
data('sleepstudy', show_doc=True)
```

```python
sns.set(style="whitegrid")

sns.lineplot(x='Days', y='Reaction', hue='Subject', data=df, alpha=0.5)

average_reaction_time = df.groupby('Days')['Reaction'].mean().reset_index()

sns.lineplot(x='Days', y='Reaction', data=average_reaction_time, color='black', linewidth=2, label='Average')

plt.xlabel('Days')
plt.ylabel('Reaction')
plt.title('Individual Subject\'s Reaction Times with Average Change')
plt.legend() 

plt.show()
```
